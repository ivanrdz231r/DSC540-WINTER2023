## 4a) Complete Activity 1 page 17

***Instructions)***

Data Wrangling with Python: Activity 1 page 17

In this activity, we will generate a list of random numbers and then generate another list from the first one, which only contains numbers that are divisible by three. Repeat the experiment three times. Then, we will calculate the average difference of length between the two lists.

These are the steps for completing this activity:

1. Create a list of 100 random numbers.
2. Create a new list from this random list, with numbers that are divisible by 3.
3. Calculate the length of these two lists and store the difference in a new variable.
4. Using a loop, perform steps 2 and 3 and find the difference variable three times.
5. Find the arithmetic mean of these three difference values.

***Answer)***

```python
# Initial list
random_number_list = [random.randint(0, 100) for i in range(0, 100)]
random_number_list
```

```python
# Divisible by 3 list
list_with_divisible_by_3 = [a for a in random_number_list if a % 3 == 0]

list_with_divisible_by_3
```

```python
# Storing the difference between both lists in a new variable
length_of_random_list = len(random_number_list)

length_of_3_divisible_list = len(list_with_divisible_by_3)

difference = length_of_random_list - length_of_3_divisible_list

difference
```

```python
# Using a loop, perform steps 2 and 3 and find the difference variable three times.
NUMBER_OF_EXPERIMENTS = 10
LIMIT = 100

difference_list = []

for i in range(0, NUMBER_OF_EXPERIMENTS):

    random_number_list = [random.randint(0, LIMIT) for x in range(0, LIMIT)]

    list_with_divisible_by_3 = [a for a in random_number_list if a % 3 == 0]

    

    length_of_random_list = len(random_number_list)

    length_of_3_divisible_list = len(list_with_divisible_by_3)

    difference = length_of_random_list - length_of_3_divisible_list

    

    difference_list.append(difference)

difference_list
```

```python
# arithmetic mean (common average) for the differences
avg_diff = sum(difference_list) / float(len(difference_list))

avg_diff
```

## 4b) Complete Activity 2 page 31

***Instructions)***

In this activity, we will do the following:

- Get multiline text and save it in a Python variable
- Get rid of all new lines in it using string methods
- Get all the unique words and their occurrences from the string
- Repeat the step to find all unique words and occurrences, without considering case sensitivity

These are the steps solve this activity

1) Create a mutliline_text variable by copying the text from the first chapter of Pride and Prejudice.
**Note**
The first chapter of Pride and Prejudice by Jane Austen has been made available on the GitHub repository at https://github.com/TrainingByPackt/Data-Wrangling-with-Python/blob/master/Chapter01/Activity02/.

2) Find the type and length of the multiline_text string using the commands type and len.
3) Remove all new lines and symbols using the replace function.
4) Find all of the words in multiline_text using the split function.
5) Create a list from this list that will contain only the unique words.
6) Count the number of times the unique word has appeared in the list using the key and value in dict.
7) Find the top 25 words from the unique words that you have found using the slice function. 

***Answer)***

```python
# Opening the txt. file

file_path = 'C:/Users/ivan2/gitLocal/DSC540-WINTER2023/P&P Ch01.txt'

with open(file_path, 'r') as file:
    multiline_text = file.read()
```

```python
# Type and length of the multiline_variable

print(type(multiline_text))

print(len(multiline_text))
```

```python
# Removing all lines
multiline_text_rpl = multiline_text.replace('\n', "")

# Removing symbols and special charicters using replace function
# Using regular expresion for a better efficiency and less code
import re

cleaned_multiline_text = re.sub(r'[?|$|.|!|"|,|;|:]',r'',multiline_text_rpl)
cleaned_multiline_text
```

```python
# finding all of the words in multiline_text using the split function

list_of_words = cleaned_multiline_text.split()
list_of_words
```

```python
len(list_of_words)
```

```python
# creating an additional list that will contain only unique words
unique_words_as_dict = dict.fromkeys(list_of_words)
len(list(unique_words_as_dict.keys()))
```

```python
# Count the number of times the unique word has appeared in the list using the key and value in dict
for word in list_of_words:
    if unique_words_as_dict[word] is None:
        unique_words_as_dict[word] = 1
    else:
        unique_words_as_dict[word] += 1
unique_words_as_dict
```

```python
# Top 25 Words

top_words = sorted(unique_words_as_dict.items(), key=lambda key_val_tuple: key_val_tuple[1], reverse=True)

top_words[:25]
```

```python
# generate all possible 3-digit numbers
permutations(range(3))
```

```python
# Loop over the iterator expression and print each element returned
for number_tuple in permutations(range(3)):
    print(number_tuple)
    assert isinstance(number_tuple, tuple)
```

```python
# Loop again with a lambda expression to drop any leading zeros from the tuples
for number_tuple in permutations(range(3)):
    print(list(dropwhile(lambda x: x <= 0, number_tuple)))
```

```python
# code for the final step of the instructions
import math
def convert_to_number(number_stack):
    final_number = 0
    for i in range(0, len(number_stack)):
        final_number += (number_stack.pop() * (math.pow(10, i)))
    return final_number
```

```python
for number_tuple in permutations(range(3)):
    number_stack = list(dropwhile(lambda x: x <= 0, number_tuple))
    print(convert_to_number(number_stack))
```

## 4d) Complete Activity 4 page 59

***Instructions)***

Design Your Own CSV Parser

These steps will guide you to solve this activity:

1) Import zip_longest from itertools. Create a function to zip header, line and fillvalue=None.
2) Open the accompanying sales_record.csv file from the GitHub link by using r mode inside a with block and first check that it is opened.
3) Read the first line and use string methods to generate a list of all the column names.
4) Start reading the file. Read it line by line.
5) Read each line and pass that line to a function, along with the list of the headers. The work of the function is to construct a dict out of these two and fill up the key:values. Keep in mind that a missing value should result in None.

***Answer)***

```python
from itertools import zip_longest
```

```python
def return_dict_from_csv_line(header, line):
    # Zip them
    zipped_line = zip_longest(header, line, fillvalue=None)
    # Use dict comprehension to generate the final dict
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict
```

```python
with open("sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")
    for i, line in enumerate(fd):
        # Here we loop over the first 10 lines in order to not make the output too big
        line = line.replace("\n", "").split(",")
        d = return_dict_from_csv_line(header, line)
        print(d)
        if i > 10:
            break
```
