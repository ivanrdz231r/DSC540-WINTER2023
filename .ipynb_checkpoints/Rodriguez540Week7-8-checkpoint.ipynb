### 1. Chapter 7 - Transform the data using a mapping dictionary

The three datasets will have to be merged. There doesn't appear to be a FK that could work across the three datasets. I will take the approach of creating a column mapping file to create a mapping dictionary. After this is done we could concat the three datasets together based on similar column names.

I created a mapping file in excel that could be used to create the mapping dictionary.

```python
# Paths to each dataset
file_location_mapping = 'C:/Users/ivan2/gitLocal/DSC540-WINTER2023/FilesWeek7-8/ColumnMappingDict.csv'

# Loading each dataset into a pandas DataFrame
df_mapping = pd.read_csv(file_location_mapping)

df_mapping.head()
```

```python
# initializing the mapping dictionary
manual_mapping_dict = {}

# populating the dictionary using "Year17" descriptions as keys
for _, row in df_mapping.iterrows():
    year17_desc = row['Year17']  # Key
    year15_desc = row['Year15']  # Value for year 15
    year16_desc = row['Year16']  # Value for year 16
    
    # ensuring that year 17 descriptions have corresponding entries from years 15 and 16
    manual_mapping_dict[year17_desc] = {
        '15': year15_desc,
        '16': year16_desc
    }

# display a subset of the manual mapping dictionary to verify its structure
list(manual_mapping_dict.items())[:10]
```

Now that I have created the mapping dictonary I need to rename the columns in df_2015 and df_2016 to match that of 2017

```python
# Rename columns in df_2015 and df_2016 using the mapping dictionary
for year17_desc, mapping in manual_mapping_dict.items():
    year15_desc = mapping.get('15')
    year16_desc = mapping.get('16')

    # if a corresponding year15 column exists, rename the column in df_2015
    if pd.notnull(year15_desc) and year15_desc in df_2015.columns:
        df_2015.rename(columns={year15_desc: year17_desc}, inplace=True)

    # if a corresponding year16 column exists, rename the column in df_2016
    if pd.notnull(year16_desc) and year16_desc in df_2016.columns:
        df_2016.rename(columns={year16_desc: year17_desc}, inplace=True)
```

### 3. Chapter 7 - Filter out missing data

*Filter out missing data*.

Most respondents that leave the first three element of question 6 blank will leave all subsequent elements and questions blank as well. For this reason we will remove records with NaN values for the first three element of Q6.

```python
# Columns to check for NaN values
nan_columns = [
    'Q6 | 100 Grand Bar',
    'Q6 | Anonymous brown globs that come in black and orange wrappers\t(a.k.a. Mary Janes)',
    'Q6 | Any full-sized candy bar'
]

# Remove records where the specified Q6 columns are all NaN
filtered_df = combined_df.dropna(subset=nan_columns, how='all')
```

### 6. Chapter 8 - Pivot "Wide to long"

Now that the data is combined and furthur cleaned, we can reshape the data to make it more managable.

```python
filtered_df.head(5)
```

```python
# adding 'Q6 |' to the old features to standardize the naming convention
filtered_df.rename(columns={
    ' [Brach products (not including candy corn)]': 'Q6 | Brach products (not including candy corn)',
    ' [Bubble Gum]': 'Q6 | Bubble Gum',
    ' [Lapel Pins]': 'Q6 | Lapel Pins',
    ' [Mary Janes]': 'Q6 | Mary Janes',
    ' [Runts]': 'Q6 | Runts', 
    ' [Mint Leaves]': 'Q6 | Mint Leaves', 
    ' [Mint M&Ms]': 'Q6 | Mint M&Ms',
    ' [Ribbon candy]': 'Q6 | Ribbon candy',
    ' [Peterson Brand Sidewalk Chalk]': 'Q6 | Peterson Brand Sidewalk Chalk',
    ' [Peanut Butter Bars]': 'Q6 | Peanut Butter Bars',
    ' [Peanut Butter Jars]': 'Q6 | Peanut Butter Jars',
    ' [Sea-salt flavored stuff, probably chocolate, since this is the "it" flavor of the year]': 'Q6 | Sea-salt flavored stuff, probably chocolate, since this is the "it" flavor of the year'
}, inplace=True)
```

```python
for column in filtered_df.columns:
    print(column)
```

```python
# Reshaping the dataset from wide to long format, focusing on the Q6 columns
df_longV1 = pd.melt(filtered_df, 
                  id_vars=[col for col in filtered_df.columns if not col.startswith('Q6')], 
                  value_vars=[col for col in filtered_df.columns if col.startswith('Q6')],
                  var_name='Candy_Type', 
                  value_name='Response')
```

The resulting `df_longV1` will have the following columns:

### 7. Chapter 10 - Grouping

I want to group the candies, view the grouped responses for each candy.

In order for this to look appropriate I first need to create dummy variables from the 'Response' feature, to get the counts of the different reponses into their own columns.

```python
# convert 'Response' into dummy variables
response_dummies = pd.get_dummies(df_longV1['Response'])

# join the dummy variables back to the original DataFrame
df_longV1 = df_longV1.join(response_dummies)
```

Now we can perform the groupby statement for the 'Candy_Type' feature.

I am specifically interested in the groupby sum() for the different candy ratings. I am not interested in the other features at this time so I will leave the other features to a default agg. of count()

```python
# define the specific aggregations for 'DESPAIR', 'JOY', and 'MEH'
aggregations = {
    'DESPAIR': 'sum',
    'JOY': 'sum',
    'MEH': 'sum'
}

# all other columns will be set to 'count'
for column in df_longV1.columns:
    if column not in aggregations:
        aggregations[column] = 'count'
```

```python
# group by 'Candy_Type' and apply the specified aggregations
grouped_df = df_longV1.groupby('Candy_Type').agg(aggregations)
```

```python
# here are the Candies with the highes Joy, just for fun.
sorted_df = grouped_df.sort_values(by='JOY', ascending=False)

sorted_df[['DESPAIR', 'JOY', 'MEH']].head(20)
```

As I expected, Reese's Peanut Butter Cups is the most popular candy, as they are my favorite!

### 8. Chapter 10 - Cross Tabs

```python
# Crosstab between 'Going Out?' and 'Year'
crosstab = pd.crosstab(df_longV1['Q1: GOING OUT?'], df_longV1['Year'], margins=True)

# Display the crosstab
print(crosstab)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# heatmap from the crosstab
plt.figure(figsize=(10, 8))  # You can adjust the figure size as needed
sns.heatmap(crosstab, annot=True, cmap="YlGnBu", fmt="d")

# labels and title
plt.xlabel('Year')
plt.ylabel('Going Out?')
plt.title('Crosstab of "Going Out?" and Response by Year')

# Display the heatmap
plt.show()
```

It looks like most people that took this survey are staying in for Halloween. I did not look into the 'Age' response yet, but these results tell me that this survey was mostly taken by adults.

Additionally fewer people took this survey in 2016 and 2017 than in 2015.
